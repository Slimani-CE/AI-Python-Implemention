```python
# Import libs
import matplotlib.pyplot as plt

# to see plots in the notebook
%matplotlib inline 

x = list(range(0, 10))
y = list(range(-10, 0))

plt.plot(x, y)
```

```python
a = [10, 20, 30, 40, 50]
b = [30, 20, 10, 20, 30]
plt.plot(a, b, c = 'r')
plt.plot(b, a, c = 'b')
plt.show()
```

```python
a = [0, -100, 25, 67, -323]
b = [0, 3, 7, 3, 9]
plt.plot(a, b, c = 'r')
plt.show()
```

```python
# Zoom in 
plt.axis([-50, 80, 2, 8])
plt.plot(a, b)
```

```python
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.axis([-50, 80, 2, 8])
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('H',"E","L","L","O", "O", "O"))
plt.plot(a, b)
```
